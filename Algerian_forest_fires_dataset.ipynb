## Algerian Forest Fires Dataset


Data Set Information:

The dataset includes 244 instances that regroup a data of two regions of Algeria, namely the Bejaia region located in the
northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region. 

The period from June 2012 to September 2012. The dataset include 11 attributes and 1 output attribute (class) The 244 instances have been classified into fire (138 classes) and not fire (106 classes) classes. 

## Attribute Information:

1. Date : (DD/MM/YYYY) Day, month ('june' to 'september'), year (2012) Weather data observations
2. Temp : temperature noon (temperature max) in Celsius degrees: 22 to 42
3. RH : Relative Humidity in %: 21 to 90
4. Ws :Wind speed in km/h: 6 to 29
5. Rain: total day in mm: 0 to 16.8 FWI Components
6. Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5
7. Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9
8. Drought Code (DC) index from the FWI system: 7 to 220.4
9. Initial Spread Index (ISI) index from the FWI system: 0 to 18.5
10. Buildup Index (BUI) index from the FWI system: 1.1 to 68
11. Fire Weather Index (FWI) Index: 0 to 31.1
12. Classes: two classes, namely Fire and not Fire

## Scatter plot to find the relationship between Temp and FWI

```python
# Selecting the 'Temperature' and 'FWI' columns
x = df['Temperature']
y = df['FWI']

# Plotting scatter plot
plt.scatter(x, y)

# Setting labels and title
plt.xlabel('Temperature (°C)')
plt.ylabel('FWI')
plt.title('Scatter Plot: Temperature vs FWI')



# Displaying the plot
plt.show()
```

```python
# selecting initial spread index as independent feature 

x = df['ISI']
y = df['FWI']

# Plotting scatter plot
plt.scatter(x, y)

# Setting labels and title
plt.xlabel('Temperature (°C)')
plt.ylabel('FWI')
plt.title('Scatter Plot: Temperature vs FWI')



# Displaying the plot
plt.show()
```

```python
# for the above scatter plot fitting the best fit line 
# Select the 'ISI' and 'FWI' columns
x = df['ISI']
y = df['FWI']

# Perform linear regression
reg = LinearRegression()
reg.fit(np.array(x).reshape(-1, 1), y)


# Get the predicted y values
y_pred = reg.predict(np.array(x).reshape(-1, 1))

# Plot scatter plot
plt.scatter(x, y)

# Plot the best-fit line
plt.plot(x, y_pred, color='red', linewidth=2)

# Set labels and title
plt.xlabel('ISI')
plt.ylabel('FWI')
plt.title('Scatter Plot: ISI vs FWI with Best-Fit Line')

# Display the plot
plt.show()
```
