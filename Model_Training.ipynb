```python
# importing the necessary library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# dropping the month, day and year column
df.drop(['day', 'month', 'year'], axis=1, inplace=True)
```

```python
# Perorming encoding because we need have many categories for same name in the Classes column 
df['Classes'].value_counts()
```

```python
df['Classes']=np.where(df['Classes'].str.contains("not fire"),0,1)
```

```python
#Differentitaing Independent and Dependent features

X=df.drop('FWI',axis=1)
y=df['FWI']
```
